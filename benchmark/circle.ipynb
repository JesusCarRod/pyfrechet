```python
import sys, os; sys.path.append(os.path.dirname(os.getcwd())) 
import numpy as np  
from pyfrechet.metric_spaces import *
from benchmark import bench

OUT_FILE = 'result_circle_cross_est_400.json'

def gen_data(N, p, alpha, beta, eps=0.1):
    M = Sphere(1)
    m = lambda x: np.power(x.dot(beta) + alpha, 2)
    x = np.random.rand(N*p).reshape((N,p))
    mx_theta = m(x)
    theta = mx_theta + eps*np.random.randn(N)
    mx = np.c_[np.cos(mx_theta), np.sin(mx_theta)]
    y = np.c_[np.cos(theta), np.sin(theta)]
    return x, MetricData(M, y), MetricData(M, mx)


bench(
    gen_data,
    OUT_FILE,
    ps=[2, 5, 10, 20],
    Ns=[50, 100, 200, 400],
    replicas=50
)
```
